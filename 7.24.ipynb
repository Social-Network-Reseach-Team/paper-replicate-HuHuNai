# 2

```r
N = 100
```

```r
help(sample)
```

```r
a=matrix(0,nrow=3,ncol=3)
a[2,3]=1
a[1,2]=1
a[1,3]=1
a
xulie=c(1,2, 1,3, 2,3)
pl <- graph(xulie, directed=F)
plot(pl, layout=layout.fruchterman.reingold)

for(i in 3 :1000){
  a=rbind(cbind(a,rep(0,i)),rep(0,i+1))#右边加一列0，底下加一行0
  x=sample(1:i,1)
  a[x,i+1]<-1 
  for(k in 1:i) { #补充左下三角值
    for(m in (k+1):(i+1)){
      a[m,k]<-a[k,m]
    }
  }
  #which(a[,i+1]!=0)
  xulie=c(xulie,i+1,which(a[,i+1]!=0)) #添加额外节点
  pl <- graph(xulie, directed=F)
  if (i%%100 == 0){
  plot(pl, layout=layout.fruchterman.reingold)      
  }
}
```

```r
degree(pl)
```

```r
plot(pl, layout=layout.fruchterman.reingold,vertex.size=1,vertex.label = NA)
V(pl)$color <- 'green'
E(pl)$color <- 'red'
plot(pl, layout=layout.fruchterman.reingold,vertex.size=1,vertex.label = NA)

plot(pl,layout=layout_as_tree,vertex.size=1,vertex.label = NA)
```

```r
df<-degree(pl) %>% 
  data.frame() %>% 
  purrr::set_names("value")

ggplot(df,aes(x=value))+
  geom_density(color="black",fill="gray")
```

```r
df_point<-table(df) %>% 
  data.frame() %>% 
  mutate(df=as.character(df)) %>% 
  mutate(df=as.numeric(df))

ggplot(df_point,aes(x=df,y=Freq))+
  geom_point()+
  geom_line()
```

```r
df_log<-table(df) %>% 
  data.frame() %>% 
  mutate(df=as.character(df)) %>% 
  mutate(df=as.numeric(df)) 

ggplot(df_log,aes(x=log(df),y=log(Freq)))+
  geom_point()+
  geom_line()
```

```r
ggplot(df_log,aes(x=df,y=log(Freq)))+
  geom_point()+
  geom_line()
```

```r
N = 1000
```

```r
pl1<-barabasi.game(N, power = 1,m=5, directed = FALSE) 
adj = get.adjacency(pl1)
sum = 0
while(sum<1000){
    x = sample(1:N,1)
    y = sample(1:N,1)
    if (adj[x,y] == 0){
        adj[x,y] <- 1
        adj[y,x] <- 1
        sum <- sum + 1
    }
}
```

```r
pl2 = from_adjacency(adj,directed = FALSE)
```
